```python
import os
import numpy as np
import matplotlib.pyplot as plt

NFEs = [5, 6, 8, 10, 12, 15, 20]
root_dirs = ['/data/guided-diffusion/scale/rbf_ecp_marginal_M=128_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_same_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_lagc_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_lagp_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_spd_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_to1_64',
             '/data/guided-diffusion/scale/rbf_ecp_marginal_to3_64',
             ]

for root_dir in root_dirs:
    print(root_dir)
    for order in [3]:
        plt.figure(figsize=[18, 3])
        for i, NFE in enumerate(NFEs):        
            npz_file = os.path.join(root_dir, f'NFE={NFE},p={order},number=0.npz')
            data = np.load(npz_file)
            plt.subplot(1, len(NFEs), i+1)
            plt.title(f"NFE={NFE}")
            plt.plot(data['optimal_log_scales'][0].T, label='predictor')
            plt.plot(data['optimal_log_scales'][1].T, label='corrector')
            plt.grid()
            plt.legend()
        plt.show()
            
```
