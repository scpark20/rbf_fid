```python
import matplotlib.pyplot as plt
import numpy as np

# NFE (x ):  [5, 6, 8, 10, 12, 15, 20]
nfe_values = np.array([5, 6, 8, 10, 12, 15, 20])

# -------------------------------------------------------
# (1) ImageNet 256, guidance scale = 2.0 ()
# -------------------------------------------------------
fid_rbf_s2      = np.array([40.2, 32.8, 24.1, 18.6, 17.2, 15.8, 14.9])
fid_solverpp_s2 = np.array([39.5, 31.5, 23.9, 18.0, 16.6, 15.3, 14.4])
fid_unipc_s2    = np.array([38.8, 30.9, 23.1, 17.4, 16.0, 14.8, 14.0])
fid_v3_s2       = np.array([37.6, 30.2, 22.5, 16.8, 15.6, 14.3, 13.7])

# -------------------------------------------------------
# (2) ImageNet 256, guidance scale = 4.0 ()
# -------------------------------------------------------
fid_rbf_s4      = np.array([42.2, 33.5, 26.1, 20.3, 17.9, 15.5, 14.7])
fid_solverpp_s4 = np.array([41.8, 32.9, 25.4, 19.8, 17.6, 15.4, 14.4])
fid_unipc_s4    = np.array([39.3, 31.2, 24.8, 18.9, 16.8, 14.9, 14.2])
fid_v3_s4       = np.array([38.0, 30.8, 23.9, 18.5, 16.2, 14.6, 13.9])

# -------------------------------------------------------
# (3) Stable-Diffusion, guidance scale = 7.5 ()
# -------------------------------------------------------
fid_rbf_sd75      = np.array([70.0, 50.2, 35.1, 23.5, 19.8, 16.2, 15.0])
fid_solverpp_sd75 = np.array([69.2, 49.0, 34.8, 23.2, 19.5, 16.0, 14.8])
fid_unipc_sd75    = np.array([68.0, 48.0, 33.5, 22.7, 19.1, 15.7, 14.5])
fid_v3_sd75       = np.array([66.8, 47.6, 32.9, 22.2, 18.8, 15.3, 14.2])

# -------------------------------------------------------
# Figure  3 Subplot 
# -------------------------------------------------------
fig, axes = plt.subplots(1, 3, figsize=(20, 5))

# ============================
# (1) ImageNet256 (s=2.0)
# ============================
axes[0].plot(nfe_values, fid_rbf_s2,      marker='o', label='RBF-Solver')
axes[0].plot(nfe_values, fid_solverpp_s2, marker='o', label='DPM-Solver++')
axes[0].plot(nfe_values, fid_unipc_s2,    marker='o', label='UniPC')
axes[0].plot(nfe_values, fid_v3_s2,       marker='o', label='DPM-Solver v3')
axes[0].set_title('ImageNet256 (s=2.0)')
axes[0].set_xlabel('NFE')
axes[0].set_ylabel('FID')
axes[0].grid(True)
axes[0].legend()

# ============================
# (2) ImageNet256 (s=4.0)
# ============================
axes[1].plot(nfe_values, fid_rbf_s4,      marker='o', label='RBF-Solver')
axes[1].plot(nfe_values, fid_solverpp_s4, marker='o', label='DPM-Solver++')
axes[1].plot(nfe_values, fid_unipc_s4,    marker='o', label='UniPC')
axes[1].plot(nfe_values, fid_v3_s4,       marker='o', label='DPM-Solver v3')
axes[1].set_title('ImageNet256 (s=4.0)')
axes[1].set_xlabel('NFE')
axes[1].set_ylabel('FID')
axes[1].grid(True)
axes[1].legend()

# ============================
# (3) Stable Diffusion (s=7.5)
# ============================
axes[2].plot(nfe_values, fid_rbf_sd75,      marker='o', label='RBF-Solver')
axes[2].plot(nfe_values, fid_solverpp_sd75, marker='o', label='DPM-Solver++')
axes[2].plot(nfe_values, fid_unipc_sd75,    marker='o', label='UniPC')
axes[2].plot(nfe_values, fid_v3_sd75,       marker='o', label='DPM-Solver v3')
axes[2].set_title('Stable-Diffusion (s=7.5)')
axes[2].set_xlabel('NFE')
axes[2].set_ylabel('FID')
axes[2].grid(True)
axes[2].legend()

plt.tight_layout()
plt.show()
```
